```python
def Factorial(num):
    
    if num==1:
        
        return 1
    
    else:
        
        fact=num*Factorial(num-1)
        
        return fact
        
```

```python
Factorial(5)
```

```python
class MATH():
    
    def __init__(self,a,b):
        
        self.a=a
        
        self.b=b
    
    def Addition(self,):
        
        return self.a+self.b
    
    def Subtraction(self,):
        
        return self.a-self.b
    
    def Division(self,):
        
        return self.a/self.b
    
    
```

```python
m=MATH(a=25,b=36)
```

```python
m.Addition()
```

```python
m.Subtraction()
```

```python
m.Division()
```
