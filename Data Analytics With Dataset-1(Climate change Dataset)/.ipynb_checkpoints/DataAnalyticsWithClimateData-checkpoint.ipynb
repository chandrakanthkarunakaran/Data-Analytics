```python
import pandas as pd
```

```python
data=pd.read_csv("GlobalLandTemperaturesByCountry.csv")
```

```python
data["dt"]=pd.to_datetime(data["dt"])
```

```python
data.loc[data["Country"]=="Antarctica","AverageTemperature"]=-45
```

```python
dataYB=data.loc[(data["dt"].dt.day==1)&(data["dt"].dt.month==1)&(data["dt"].dt.year>1950)]
```

```python
dataYB["AverageTemperature"].fillna(method="ffill",inplace=True)
```

Temperature Trends Comparison Between India and Pakistan

```python
dataIndia=dataYB.loc[dataYB["Country"]=="India"]
```

```python
dataPK=dataYB.loc[dataYB["Country"]=="Pakistan"]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dataIndia["dt"],dataIndia["AverageTemperature"],marker="o")
plt.plot(dataPK["dt"],dataPK["AverageTemperature"],marker="x")
plt.title("Average Temperature Comparison Between India vs Pakistan")
plt.legend(["India","Pakistan"])
```

Temperature Comparison Between India And Taiwan

```python
def cm_to_inch(value):
    return value/2.54
```

```python
dataIndia=dataYB.loc[dataYB["Country"]=="India"]
dataTaiwan=dataYB.loc[dataYB["Country"]=="Taiwan"]

plt.figure(figsize=(cm_to_inch(15),cm_to_inch(10)))

plt.plot(dataIndia["dt"],dataIndia["AverageTemperature"],marker="o",color="b")
plt.plot(dataTaiwan["dt"],dataTaiwan["AverageTemperature"],marker="o",color="r",alpha=0.5)
plt.title("Average Temperature Comparison Between India and Taiwan (1950-2015)")
plt.legend(["India","Taiwan"])
```

-Plotting Average Temperatures Across Globe From 1950-2015

```python
countryGrps=dataYB.groupby("Country")
```

```python
avgTempCount={"Country":[],"AverageTemperature":[]}

for grp in countryGrps.groups:

    averageTempCountry=countryGrps.get_group(grp)["AverageTemperature"].mean()

    avgTempCount["Country"].append(grp)

    avgTempCount["AverageTemperature"].append(averageTempCountry)


avgTempCountDF=pd.DataFrame(avgTempCount)



```

```python
avgTempCountDF.sort_values(by="AverageTemperature",ascending=False,inplace=True)
```

```python
avgTempCountDF.reset_index(inplace=True,drop=True)
```

```python
avgTempCountDF.loc[avgTempCountDF["Country"]=="Antarctica"]
```

```python
import geopandas as gpd
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df_world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
# Creating Mapping For Similar Names Beween geopandas data and avgTemp DataFrame

cntNotInWorld=[x for x in list(set(avgTempCountDF.Country)) if x not in list(set(df_world.name))]
wldNotInCnt=[x for x in list(set(df_world.name)) if x not in list(set(avgTempCountDF.Country))]

namesSimilar={}

for country in wldNotInCnt:

    tokens1=country.lower().split(" ")

    for c in cntNotInWorld:

        tokens2=c.lower().split(" ")

        tokens2=[x for x in tokens2 if x not in ["states","of","islands","island","north","south"]]

        matchToken=[x for x in tokens1 if x in tokens2]

        if len(matchToken)/len(tokens2)>0.4:

            namesSimilar[c]=country
            
avgTempCountDF["Country"].replace(to_replace=namesSimilar,inplace=True)
```

```python
df_world=df_world.merge(avgTempCountDF, how="left", left_on=['name'], right_on=['Country'])
```

```python
ax = df_world["geometry"].boundary.plot(figsize=(20,16))

df_world.plot(column="AverageTemperature", ax=ax, cmap='OrRd', 
                     legend=True,legend_kwds={"label": "Average Temperature", "orientation":"horizontal"},
                     missing_kwds={'color': 'grey'})
ax.set_title("Average Temperature Across World From 1950-2015")
```

Using Logistic Regression to Predict Temperature Trend daywise

```python
dataIndia=data.loc[data["Country"]=="India"]
```

```python
dataIndia.AverageTemperature.fillna(method="ffill",inplace=True)
```

```python
dataIndia["Month"]=dataIndia["dt"].dt.month
dataIndia["Day"]=dataIndia["dt"].dt.day
```

```python
dataIndia["Year"]=dataIndia.dt.dt.year
```

```python
dataIndia
```

```python
dataX=dataIndia[["Month","Year"]]
dataY=dataIndia["AverageTemperature"]
```

```python
# Find Correlation between Month and Average Temperature
```

```python
dataIndia.columns
```

```python
import seaborn as sns
```

```python
sns.set(style="ticks", color_codes=True)    
g = sns.pairplot(dataIndia)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(dataX,dataY,test_size=0.2,random_state=21)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression().fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
meanError=mean_absolute_error(list(y_test),list(y_pred))
```
