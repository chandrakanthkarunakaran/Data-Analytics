MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data=pd.read_csv("CROPSINDIA2002-2022.csv")
```

```python
data.fillna(method='ffill',inplace=True)
```

```python
dataStates=data.loc[~data["State / Union Territory"].isin(['All India',
 'All States',
 'Union Territories'])]
```

```python
dataStates.loc[:,"State"]=[x.split(".")[-1].strip() for x in dataStates["State / Union Territory"]]
```

```python
dataStates
```

```python
numCols=['Rice', 'Wheat', 'Coarse cereals',
       'Pulses', 'Food- grains', 'Oilseeds', 'Cotton (Lint)', 'Sugar- cane',
       'Raw Jute & Mesta']

dataStates[numCols]=dataStates[numCols].replace(to_replace="-",value=0)
```

```python



for col in numCols:
    
    print(col)
    
    dataStates[col]=dataStates[col].astype(float)
```

```python
dataAP=dataStates.loc[dataStates["State"]=="Andhra Pradesh"]
```

```python
xAxis=dataAP["Year"]
yAxis=dataAP["Rice"]
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
ax.plot(xAxis, yAxis)
plt.xticks(rotation=270)
plt.title("Rice Production In Andhra Pradesh 2002-2022")
```

```python
# Multiple Lines Rice vs Sugarcane in Andhra Pradesh

y2Axis=dataAP["Sugar- cane"]

```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
ax.plot(xAxis, yAxis,marker="8",markersize=10,color="green")
ax.plot(xAxis,y2Axis,marker="*",markersize=10,color="brown")
plt.xticks(rotation=270)
plt.title("Rice Vs Sugar-Cane Production In Andhra Pradesh 2002-2022")
plt.legend(["Rice","Sugar-Cane"])
```

```python
states=list(set(dataStates.State))
```

```python
states
```

```python
# Rice Production Trend Across All STates
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)


for state in states:
    
    xAxis=dataStates.loc[dataStates["State"]==state,"Year"]
    
    yAxis=dataStates.loc[dataStates["State"]==state,"Rice"]
    
    ax.plot(xAxis, yAxis,marker="8",markersize=5)

plt.xticks(rotation=270)
plt.title("Rice Vs Sugar-Cane Production All States 2002-2022")
#plt.legend(states)
```

Bar Chart

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.bar(xAxis,yAxis,color="#800835",edgecolor="black",label="Rice",)
ax.bar(xAxis,y2Axis,bottom=yAxis,edgecolor="black",label="wheat")

ax.legend(loc="upper left")

plt.xticks(rotation=270)
plt.title("Rice Production In Andhra Pradesh (2002-2022)")

plt.show()
```

```python
# Horizontal Bars

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.barh(xAxis,yAxis,color="brown",edgecolor="black")
plt.xticks(rotation=270)
plt.title("Rice Production In Andhra Pradesh (2002-2022)")

plt.show()
```
