Joining

* Inner Join
* Left Join
* Right Join

```python
dataTxn
```

```python
txnDF=txnDF.append(pd.DataFrame({"Transaction ID":["TR10"],"Item":["Poori"],"Qty":[3],"Rate":[25],"Amount":[75]}))
```

```python
pd.merge(left=txnDF,right=dataTxn["DiscPerc"],how="right",left_on="Item",right_on="Item")
```

```python
pd.merge(left=txnDF,right=dataTxn["DiscPerc"],how="left",left_on="Item",right_on="Item")
```

```python
pd.merge(left=txnDF,right=dataTxn["DiscPerc"],how="inner",left_on="Item",right_on="Item")
```

```python
pd.merge(left=txnDF,right=dataTxn["DiscPerc"],how="outer",left_on="Item",right_on="Item")
```

Pivoting

```python
data=pd.read_csv("CROPSINDIA2002-2022.csv")
```

```python
data.fillna(method='ffill',inplace=True)
```

```python
dataL1=data.loc[~data["State / Union Territory"].isin(['All India',
 'All States',
 'Union Territories'])]
```

```python
dataL1.columns
```

```python
pd.pivot_table(data=dataL1,values=["Rice"],index="Year",columns="State / Union Territory",aggfunc=sum)
```

```python
import numpy as np
```

```python
dataL1[numCols]=dataL1[numCols].replace(to_replace="-",value=0)
```

```python
numCols=['Rice', 'Wheat', 'Coarse cereals',
       'Pulses', 'Food- grains', 'Oilseeds', 'Cotton (Lint)', 'Sugar- cane',
       'Raw Jute & Mesta']


for col in numCols:
    
    print(col)
    
    dataL1[col]=dataL1[col].astype(float)
```

```python
pd.pivot_table(data=dataL1,values=numCols,index="State / Union Territory",aggfunc=sum)
```

Mathematical Operations

```python
dataAP=dataL1.loc[data["State / Union Territory"]=="1. Andhra Pradesh"]
```

```python
dataAP["Rice"]+dataAP["Wheat"]
```

```python
dataAP["State / Union Territory"]+"/"+dataAP["Year"]
```

```python
dataAP["StartYear"]=[int(x.split("-")[0]) for x in dataAP["Year"]]
dataAP["EndYear"]=[int(x.split("-")[1]) for x in dataAP["Year"]]

```

```python
pd.pivot_table(dataAP.loc[dataAP.EndYear<=10],values=numCols,index="State / Union Territory",aggfunc=sum)
```

```python
dataL1["StartYear"]=[int(x.split("-")[0]) for x in dataL1["Year"]]
dataL1["EndYear"]=[int(x.split("-")[1]) for x in dataL1["Year"]]

```

```python
data200210=pd.pivot_table(dataL1.loc[dataL1.EndYear<=10],values=numCols,index="State / Union Territory",aggfunc=sum)
```
