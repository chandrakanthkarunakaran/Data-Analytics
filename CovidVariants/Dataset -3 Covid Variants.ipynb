```python
import pandas as pd
```

```python
data=pd.read_csv("covid-variants.csv")
```

```python
data=data.loc[(data["num_sequences"]!=0)&(~data["variant"].isin(["others","non_who"]))]
```

```python
data.variant.unique()
```

```python
variantGrps=data.groupby("variant")
```

```python
data.loc[:,"date"]=pd.to_datetime(data["date"])
```

```python
variantDurList=[]

for variant in variantGrps.groups:
    
    variantDF=variantGrps.get_group(variant)
    
    obj={"Variant":variant,"StartDate":variantDF['date'].min(),
         "EndDate":variantDF["date"].max(),"Duration":(variantDF["date"].max()-variantDF["date"].min()).days}
    
    variantDurList.append(obj)
    
```

```python
variantsDF=pd.DataFrame.from_dict(variantDurList,orient="columns")
```
